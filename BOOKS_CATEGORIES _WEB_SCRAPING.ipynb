```python
def rate_num(rate):
    if rate==' One':
        return 1
    elif rate=='Two':
        return 2
    elif rate=='Three':
        return 3 
    elif rate=='Four':
        return 4
    elif rate=='Five':
        return 5
```

# BOOKS AND ITS CATEGORIES  INSERTED IN DB BROWSER



```python
conn=sqlite3.connect('ALL_BOOKS.db')

response=requests.get('https://books.toscrape.com')

soup1 = BeautifulSoup(response.text, 'html.parser')
x=soup1.find('div',attrs={'class','side_categories'}).find('ul').find('li').find('ul').find_all('a')
for b in x:
    link='https://books.toscrape.com/'+b.get('href')
    name = b.get_text().strip()
    
    sql1='INSERT INTO categories2 (cat_name) VALUES(?)'
    
    categories_data=[name]
    cursor=conn.execute(sql1,categories_data)
    cat_id=cursor.lastrowid
    conn.commit()
    
     
    # cat_id =b lastrowid
    
    response=requests.get(link)
    print(name, link) 
    
    soup2 = BeautifulSoup(response.text, 'html.parser')
    
    for article in soup2.find_all('article'):
        
        Name=article.find('h3').find('a').get('title')
    
    
            
        Price=article.find('p',attrs={'class':'price_color'}).text[2:]

            
        rate=article.find('p',attrs={'class':'star-rating'}).get('class')[1]
            
        rate_int=rate_num(rate)
        sql2='INSERT INTO books2 (name ,price,rate,category_id) VALUES(?,?,?,?)'
    
        books_data=[Name,Price,rate_int,cat_id]
        cursor=conn.execute(sql2,books_data)
        conn.commit()
        
        # insert book in db
            
        print(f"title {Name}, price : {Price}, rate {rate_int}")
             
    print('-'*80)
conn.close()
    


```
